```python
#multidimensional2d
import math
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

n1=int(input("enter quantum number n1"))
n2=int(input("enter quantum number n2"))


ax=plt.axes(projection="3d")
a=np.linspace(0,3.14,100)
b=np.linspace(0,3.14,100)
x,y= np.meshgrid(a,b)
z = (np.sqrt(4/(2*2)))*(np.sin(n1 * 3.14 * x / 2))*(np.sin( n2* 3.14 * y / 2 ))

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('\u03C8')
ax.plot_surface(x,y,z)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#initialise
v_0 = 36
E=np.linspace(-v_0,0,10000)

f=np.sqrt(E+v_0)
g=np.sqrt(-E)

#we define the transcendental functions
r=g
l1=f*np.tan(f)
l2=-f/np.tan(f+1e-9)

#plot curves
plt.figure(figsize=(6,3))
plt.ylim(-10,10)
plt.plot(E, l1, label = 'v.tanv')
plt.plot(E, l2, label = '-v.cotv')
plt.plot(E,r)
plt.legend()
plt.show()

```
